```python
# imports
import json
import pandas as pd
```

# EDA - User Data

```python
# import json data
users = []
for line in open('users.json', 'r'):
    users.append(json.loads(line))
    
# put data into dataFrame
users = pd.DataFrame(users)
```

```python
# check head of dataFrame
users.head()
```

```python
# summary of dataFrame

users.info()
```

### Takeaways

    1. There are 495 user IDs provided. Are all unique or any duplicates?
    2. There is an extra column (signUpSource) that was not provided in the data schema.

## EDA of _id column

```python
# view column names of dataframe
users.columns
```

```python
# Return the rows with duplicate user IDs and put in new dataframe
dupl_users = users.loc[users['_id'].duplicated(),:]

# view dataframe
dupl_users
```

```python
# Looks like there is more than one role. how many unique roles?

dupl_users['role'].nunique()
```

### User Data Takeaways
1. There are duplicate user IDs (_id column) in the users data. 
    
    _id is the primary key and values in this column should be unique. Looks like the duplicate user IDs have the same creation date. Maybe they were all created by accident on the same day by the same user? 
     
     
2. There is an additional role ('fetch-staff') in the role column that wasn't mentioned in User data schema. 
    
    Some of the duplicate user IDs were created by fetch-staff. Maybe test accounts?
    
    
3. There is an additional column ('signUpSource' column) that was not mentioned in the User Data Schema.

```python
# imports
import json
import pandas as pd
```

# EDA - User Data

```python
# import json data
users = []
for line in open('users.json', 'r'):
    users.append(json.loads(line))
    
# put data into dataframe
users = pd.DataFrame(users)
```

## EDA of _id column

```python
users.columns
```

```python
# Return the rows with duplicate user IDs
dupl_users = users.loc[users['_id'].duplicated(),:]
dupl_users
```

### Takeaways
1. There are duplicate user IDs in the users data. _id is the primary key and values in this column should be unique. 
2. Looks like the duplicate user IDs have the same creation date. Maybe they were all created on similar dates.
3. Some of the duplicate user IDs were created by fetch-staff. Maybe test accounts?
